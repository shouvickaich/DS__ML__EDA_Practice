```python
data["payment"].fillna(data["payment"].mode()[0],inplace=True)
data.isnull().sum()
```

```python
data["dropoff_zone"].fillna(data["dropoff_zone"].mode()[0],inplace=True)
data["pickup_zone"].fillna(data["pickup_zone"].mode()[0],inplace=True)
data[["pickup_zone"]].fillna(data["pickup_zone"].mode()[0],inplace=True)
data["pickup_borough"].fillna(data["pickup_borough"].mode()[0],inplace=True)
data["dropoff_borough"].fillna(data["dropoff_borough"].mode()[0],inplace=True)
data.isnull().sum()
```

```python
data.describe()

```

```python
# outlier detection in "passengers"
Q1_passaenger_count = data["passengers"].quantile(0.25)
Q3_passaenger_count = data["passengers"].quantile(0.75)
IQR_passaenger_count = Q3_passaenger_count - Q1_passaenger_count
outerlire_passaenger_count = data[(data["passengers"] < Q1_passaenger_count - 1.5 * IQR_passaenger_count) | (data["passengers"] > Q3_passaenger_count + 1.5 * IQR_passaenger_count)]
outerlire_passaenger_count
```

```python
# outlier detection in "distance"
Q1_distance = data["distance"].quantile(0.25)
Q3_distance = data["distance"].quantile(0.75)
IQR_distance = Q3_distance - Q1_distance
outerlire_distance = data[(data["distance"] < Q1_distance - 1.5 * IQR_distance) | (data["distance"] > Q3_distance + 1.5 * IQR_distance)]
outerlire_distance  
```

```python
# outlier detection in "fare"
Q1_fare = data["fare"].quantile(0.25)
Q3_fare = data["fare"].quantile(0.75)
IQR_fare = Q3_fare - Q1_fare
outerlire_fare = data[(data["fare"] < Q1_fare - 1.5 * IQR_fare) | (data["fare"] > Q3_fare + 1.5 * IQR_fare)]
outerlire_fare  
```

```python
# outlier detection in "tip"
Q1_tip = data["tip"].quantile(0.25)
Q3_tip = data["tip"].quantile(0.75)
IQR_tip = Q3_tip - Q1_tip
outerlire_tip = data[(data["tip"] < Q1_tip - 1.5 * IQR_tip) | (data["tip"] > Q3_tip + 1.5 * IQR_tip)]
outerlire_tip  
```

```python
# outlier detection in "tolls"
Q1_tolls = data["tolls"].quantile(0.25)
Q3_tolls = data["tolls"].quantile(0.75)
IQR_tolls = Q3_tolls - Q1_tolls
outerlire_tolls = data[(data["tolls"] < Q1_tolls - 1.5 * IQR_tolls) | (data["tolls"] > Q3_tolls + 1.5 * IQR_tolls)]
outerlire_tolls  
```

```python
# outlier detection in "total"
Q1_total = data["total"].quantile(0.25)
Q3_total = data["total"].quantile(0.75)
IQR_total = Q3_total - Q1_total
outerlire_total = data[(data["total"] < Q1_total - 1.5 * IQR_total) | (data["total"] > Q3_total + 1.5 * IQR_total)]
outerlire_total  
```
